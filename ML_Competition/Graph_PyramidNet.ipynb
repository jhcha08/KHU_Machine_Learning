# 옵티마이저 변화

```python
import matplotlib.pyplot as plt
import numpy as np

# PyramidNet_bs64_lr0.001
k = [0.1,1.1,2.1,3.1]
train = [0.87337, 0.65526, 0.95996, 0.96001]
test = [0.78198, 0.65237, 0.81147, 0.81637]

label = ['SGD','AdaGrad','RMSProp', 'Adam']

plt.figure()

x = np.arange(len(label))

plt.bar(x-0.1, train, label='Train', width = 0.2, color='lightsalmon')
plt.bar(x+0.1, test, label='Test', width = 0.2, color='navajowhite')
plt.xticks(x, label)
plt.ylim(0,1.199)

for i, v in enumerate(k):
    plt.text(v, test[i], test[i],
             fontsize = 9, 
             color='blue',
             horizontalalignment='center',
             verticalalignment='bottom')

plt.legend()
plt.xlabel('Optimizer')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy with Changing Optimizer')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

SAVEPATH = 'pyramid_opt\\'

text_acc1 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_Adagrad_acc.csv')
text_acc2 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_Adam_acc.csv')
text_acc3 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_RMSprop_acc.csv') # 내용없음
text_acc4 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_SGD_acc.csv')

text_loss1 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_Adagrad_loss.csv')
text_loss2 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_Adam_loss.csv')
text_loss3 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_RMSprop_loss.csv')
text_loss4 = pd.read_csv(SAVEPATH+'PyramidNet_bs64_lr0.001_SGD_loss.csv')

epochs = list(text_acc1['Epoch'])

# 하이퍼 파라미터에 따른 Train Acc 변화 그래프

acc1 = list(text_acc1['Train Acc'])
acc2 = list(text_acc2['Train Acc'])
acc3 = list(text_acc3['Train Acc'])
acc4 = list(text_acc4['Train Acc'])

plt.plot(epochs, acc1,label='AdaGrad', color='springgreen')
plt.plot(epochs, acc2,label='Adam', color='violet')
plt.plot(epochs, acc3,label='RMSProp', color='dodgerblue')
plt.plot(epochs, acc4,label='SGD', color='chocolate')

plt.title('Train Accuracy with Changing Optmizer')
plt.legend()
plt.show()

# 하이퍼 파라미터에 따른 Train Loss 변화 그래프

loss1 = list(text_loss1['Train Loss'])
loss2 = list(text_loss2['Train Loss'])
loss3 = list(text_loss3['Train Loss'])
loss4 = list(text_loss4['Train Loss'])

plt.plot(epochs, loss1,label='AdaGrad', color='springgreen')
plt.plot(epochs, loss2,label='Adam', color='violet')
plt.plot(epochs, loss3,label='RMSProp', color='dodgerblue')
plt.plot(epochs, loss4,label='SGD', color='chocolate')

plt.title('Train Loss with Changing Optmizer')
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# PyramidNet_Adam_bs64
k = [0.1,1.1,2.1,3.1]
train = [0.91063, 0.95626, 0.96001, 0.90336]
test = [0.80182, 0.81471, 0.81637, 0.76725]

label = ['0.1','0.01','0.001','0.0001']

plt.figure()

x = np.arange(len(label))

plt.bar(x-0.1, train, label='Train', width = 0.2, color='lightsalmon')
plt.bar(x+0.1, test, label='Test', width = 0.2, color='navajowhite')
plt.xticks(x, label)
plt.ylim(0,1.17)

for i, v in enumerate(k):
    plt.text(v, test[i], test[i],
             fontsize = 9, 
             color='blue',
             horizontalalignment='center',
             verticalalignment='bottom')

plt.legend()
plt.xlabel('Learning Rate')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy with Changing Learning Rate')
plt.show()

plt.savefig('PyramidNet_lr.jpg')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

SAVEPATH = 'pyramid_lr\\'

text_acc1 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.0001_acc.csv')
text_acc2 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.001_acc.csv')
text_acc3 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.01_acc.csv')
text_acc4 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.1_acc.csv')

text_loss1 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.0001_loss.csv')
text_loss2 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.001_loss.csv')
text_loss3 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.01_loss.csv')
text_loss4 = pd.read_csv(SAVEPATH+'PyramidNet_Adam_bs64_lr0.1_loss.csv')

epochs = list(text_acc1['Epoch'])

# 하이퍼 파라미터에 따른 Train Acc 변화 그래프

acc1 = list(text_acc1['Train Acc'])
acc2 = list(text_acc2['Train Acc'])
acc3 = list(text_acc3['Train Acc'])
acc4 = list(text_acc4['Train Acc'])

plt.plot(epochs, acc1,label='0.0001', color='springgreen')
plt.plot(epochs, acc2,label='0.001', color='violet')
plt.plot(epochs, acc3,label='0.01', color='dodgerblue')
plt.plot(epochs, acc4,label='0.1', color='chocolate')

plt.title('Train Accuracy with Changing Learning rate')
plt.legend()
plt.show()

# 하이퍼 파라미터에 따른 Train Loss 변화 그래프

loss1 = list(text_loss1['Train Loss'])
loss2 = list(text_loss2['Train Loss'])
loss3 = list(text_loss3['Train Loss'])
loss4 = list(text_loss4['Train Loss'])

plt.plot(epochs, loss1,label='0.0001', color='springgreen')
plt.plot(epochs, loss2,label='0.001', color='violet')
plt.plot(epochs, loss3,label='0.01', color='dodgerblue')
plt.plot(epochs, loss4,label='0.1', color='chocolate')

plt.title('Train Loss with Changing Learning rate')
plt.legend()

plt.show()
```
