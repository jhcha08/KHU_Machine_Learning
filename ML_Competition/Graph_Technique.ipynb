# 모델간 비교 (bs64, lr 0.001, Adam)

```python
import matplotlib.pyplot as plt
import numpy as np

k = [0.1,1.1,2.1]
train = [0.75009, 0.83932, 0.96001]
test = [0.70486,0.73293, 0.81637]

label = ['DenseNet', 'Xception', 'PyramidNet']

plt.figure()

x = np.arange(len(label))

plt.bar(x-0.1, train, label='Train', width = 0.2, color = 'mediumpurple')
plt.bar(x+0.1, test, label='Test', width = 0.2, color = 'navy')
plt.xticks(x, label)
plt.ylim(0,1.2)

for i, v in enumerate(k):
    plt.text(v, test[i], test[i],
             fontsize = 9, 
             color='blue',
             horizontalalignment='center',
             verticalalignment='bottom')

plt.legend()
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy with Changing Model')
plt.show()
```
