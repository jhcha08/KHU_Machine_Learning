# 옵티마이저 변화

```python
import matplotlib.pyplot as plt
import numpy as np

# DenseNet_bs64_lr0.001

train = [0.64483, 0.63807, 0.77994, 0.75009]
test = [0.62461, 0.61993, 0.69332, 0.70486]

label = ['SGD','AdaGrad','RMSProp', 'Adam']

plt.figure()

x = np.arange(len(label))

plt.bar(x-0.1, train, label='Train', width = 0.2)
plt.bar(x+0.1, test, label='Test', width = 0.2)
plt.xticks(x, label)
plt.ylim(0,1.17)

plt.legend()
plt.xlabel('Optimizer')
plt.ylabel('Accuracy')
plt.title('Train and Test Accuracy with Changing Optimizer')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

SAVEPATH = 'densenet_opt\\'

text_acc1 = pd.read_csv(SAVEPATH+'densenet_acc_train_adagrad_64_0.001.csv')
text_acc2 = pd.read_csv(SAVEPATH+'densenet_acc_train_adam_64_0.001.csv')
text_acc3 = pd.read_csv(SAVEPATH+'densenet_acc_train_rmsprop_64_0.001.csv')
text_acc4 = pd.read_csv(SAVEPATH+'densenet_acc_train_sgd_64_0.001.csv')

text_loss1 = pd.read_csv(SAVEPATH+'densenet_loss_train_adagrad_64_0.001.csv')
text_loss2 = pd.read_csv(SAVEPATH+'densenet_loss_train_adam_64_0.001.csv')
text_loss3 = pd.read_csv(SAVEPATH+'densenet_loss_train_rmsprop_64_0.001.csv')
text_loss4 = pd.read_csv(SAVEPATH+'densenet_loss_train_sgd_64_0.001.csv')

epochs = list(text_acc1['Epoch'])

# 하이퍼 파라미터에 따른 Train Acc 변화 그래프

acc1 = list(text_acc1['Train Acc'])
acc2 = list(text_acc2['Train Acc'])
acc3 = list(text_acc3['Train Acc'])
acc4 = list(text_acc4['Train Acc'])

plt.plot(epochs, acc1,label='AdaGrad', color='springgreen')
plt.plot(epochs, acc2,label='Adam', color='violet')
plt.plot(epochs, acc3,label='RMSProp', color='dodgerblue')
plt.plot(epochs, acc4,label='SGD', color='chocolate')

plt.title('Train Accuracy with Changing Optmizer')
plt.legend()
plt.show()

# 하이퍼 파라미터에 따른 Train Loss 변화 그래프

loss1 = list(text_loss1['Train Loss'])
loss2 = list(text_loss2['Train Loss'])
loss3 = list(text_loss3['Train Loss'])
loss4 = list(text_loss4['Train Loss'])

plt.plot(epochs, loss1,label='AdaGrad', color='springgreen')
plt.plot(epochs, loss2,label='Adam', color='violet')
plt.plot(epochs, loss3,label='RMSProp', color='dodgerblue')
plt.plot(epochs, loss4,label='SGD', color='chocolate')

plt.title('Train Loss with Changing Optmizer')
plt.legend()

plt.show()
```
